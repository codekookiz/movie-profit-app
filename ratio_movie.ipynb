```python
import pandas as pd, numpy as np, matplotlib.pyplot as plt, seaborn as sb
```

```python
df = pd.read_csv('data/new_movie.csv')
df
```

```python
df = df.loc[df['domestic_gross'] != 0, :]
```

```python
df['world/domestic ratio'] = (df['worldwide_gross'] / df['domestic_gross']).round(1)
```
