```python
df = pd.read_csv('../data/new_movie.csv')
df
```

```python
df = df.loc[df['북미 박스오피스 수익 ($)'] != 0, :]
```

```python
df['전세계/북미 수익 비율'] = (df['전세계 박스오피스 수익 ($)'] / df['북미 박스오피스 수익 ($)']).round(1)
```
